```python
import git
!git branch
!git status
import telebot
import webbrowser
from telebot import types
import sqlite3

TOKEN = '7772114816:AAGMW8N6k4iY6s6U0-sIsODjdBsztzhKr2w'

bot = telebot.TeleBot('7772114816:AAGMW8N6k4iY6s6U0-sIsODjdBsztzhKr2w')

name = None

subjects = {
    "История",
    "Русский язык",
    "Английский язык",
    "Математика",
    "География",
    "Физика",
    "Биология",
    "Химия",
    "Информатика",
    "ОБЗР",
    "Литература",
    "Труд",
    "Обществознание"
}

# Учителя
teachers = {
    "История": ["Ольга Викторовна", "Елена Александровна"],
    "Русский язык": ["Юлия Викторовна", "Мария Павловна", "Галина Дмитриевна"],
    "Математика": ["Светлана Владимировна", "Юлия Александровна", "Анастасия Сергеевна", "Ирина Ивановна"],
    "Физика": ["Светлана Владимировна", "Алена Игоревна"],
    "Английский язык": ["Наталья Олеговна", "Ирина Николаевна", "Надежда Петровна", "Надежда Алексеевна"],
    "Биология": ["Виктория Сергеевна", "Татьяна Юревна"],  # Исправлено "Юревьана" -> "Юревна"
    "ОБЗР": ["Алексей Юрьевич", "Денис Валерьевич"],
    "Информатика": ["Александра Ивановна", "Вячеслав Викторович"],
    "География": ["Татьяна Юревна"],
    "Химия": ["Николай Евгеньевич"],
    "Труд": ["Роман Викторович"],
    "Литература": ["Юлия Викторовна", "Мария Павловна", "Галина Дмитриевна"],
    "Обществознание": ["Елена Александровна", "Ольга Викторовна"]
}

# Кабинеты
cabinets = {
    "Алена Игоревна": "Кабинет 510",
    "Александра Ивановна": "Кабинет 109",
    "Юлия Викторовна": "Кабинет 410",
    "Мария Павловна": "Кабинет 305",
    "Галина Дмитриевна": "Кабинет 308",
    "Светлана Владимировна": "Кабинет 512",
    "Юлия Александровна": "Кабинет 401",
    "Анастасия Сергеевна": "Кабинет 404",
    "Виктория Сергеевна": "Кабинет 510 или 502",
    "Алексей Юрьевич": "Кабинет 307",
    "Вячеслав Викторович": "Кабинет 405",
    "Елена Александровна": "501",
    "Ольга Викторовна": "409",
    "Роман Викторович": "502",
    "Николай Евгеньевич": "510",
    "Татьяна Юревна": "510",
    "Денис Валерьевич": "410 или 510",
    "Надежда Петровна": "311",
    "Наталья Олеговна": "209",
    "Ирина Николаевна": "307",
    "Надежда Алексеевна": "412",
    "Ирина Ивановна": "402"
}

# 📌 Функция для меню (чтобы не дублировать код)
def subjects_menu():
    markup = types.InlineKeyboardMarkup()

    markup.row(
        types.InlineKeyboardButton("История", callback_data="История"),
        types.InlineKeyboardButton("Русский язык", callback_data="Русский язык")
    )
    markup.row(
        types.InlineKeyboardButton("Английский язык", callback_data="Английский язык"),
        types.InlineKeyboardButton("Математика", callback_data="Математика")
    )
    markup.row(
        types.InlineKeyboardButton("География", callback_data="География"),
        types.InlineKeyboardButton("Физика", callback_data="Физика")
    )
    markup.row(
        types.InlineKeyboardButton("Биология", callback_data="Биология"),
        types.InlineKeyboardButton("Химия", callback_data="Химия")
    )
    markup.row(
        types.InlineKeyboardButton("Информатика", callback_data="Информатика"),
        types.InlineKeyboardButton("ОБЗР", callback_data="ОБЗР")
    )
    markup.row(
        types.InlineKeyboardButton("Литература", callback_data="Литература"),
        types.InlineKeyboardButton("Труд", callback_data="Труд"),
        types.InlineKeyboardButton("Обществознание", callback_data="Обществознание")
    )
    markup.row(types.InlineKeyboardButton("📚 Шпаргалки", callback_data="cheats"))

    return markup

# ▶️ СТАРТ
@bot.message_handler(commands=['start'])
def start(message):
    user_name = message.from_user.first_name  # имя пользователя

    bot.send_message(
        message.chat.id,
        f"👋 Привет, {user_name}!\n\nВыбери предмет:",
        reply_markup=subjects_menu()
    )

# 🔘 ОБРАБОТКА КНОПОК
@bot.callback_query_handler(func=lambda call: True)
def callback(call):
    data = call.data

    # 📚 Шпаргалки
    if data == "cheats":
        markup = types.InlineKeyboardMarkup()
        markup.row(types.InlineKeyboardButton("🌐 Открыть сайт", url='https://webium.ru/media/material/'))
        markup.row(types.InlineKeyboardButton("⬅️ Назад", callback_data="back"))

        bot.edit_message_text(
            f"📚 Шпаргалки:\n\n{'https://webium.ru/media/material/'}",
            chat_id=call.message.chat.id,
            message_id=call.message.message_id,
            reply_markup=markup
        )

    # 📘 Предмет
    elif data in teachers:
        markup = types.InlineKeyboardMarkup()

        for teacher in teachers[data]:
            markup.row(types.InlineKeyboardButton(f"👨‍🏫 {teacher}", callback_data=teacher))

        markup.row(types.InlineKeyboardButton("⬅️ Назад", callback_data="back"))

        bot.edit_message_text(
            f"📘 {data}\nВыберите преподавателя:",
            chat_id=call.message.chat.id,
            message_id=call.message.message_id,
            reply_markup=markup
        )

    # 👨‍🏫 Учитель
    elif data in cabinets:
        markup = types.InlineKeyboardMarkup()
        markup.row(types.InlineKeyboardButton("⬅️ Назад", callback_data="back"))

        bot.edit_message_text(
            f"👨‍🏫 {data}\n📍 {cabinets[data]}",
            chat_id=call.message.chat.id,
            message_id=call.message.message_id,
            reply_markup=markup
        )

    # 🔙 Назад
    elif data == "back":
        bot.edit_message_text(
            "Выбери предмет:",
            chat_id=call.message.chat.id,
            message_id=call.message.message_id,
            reply_markup=subjects_menu()
        )

    # ⚠️ Ошибка
    else:
        bot.answer_callback_query(call.id, "Ошибка выбора")

bot.polling(none_stop=True)
```
